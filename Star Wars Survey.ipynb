```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
print (star_wars)
```

## Cleaning And Mapping Yes/No Columns

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()

```

```python
yes_no = {
    "Yes": True,
    "No": False
}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

## Cleaning And Mapping Checkbox Columns

#### Changing survey response items of movies seen to True or False

```python
movies_seen = {np.nan: False,
               "Star Wars: Episode I  The Phantom Menace": True,
               "Star Wars: Episode II  Attack of the Clones": True,
               "Star Wars: Episode III  Revenge of the Sith": True,
               "Star Wars: Episode IV  A New Hope": True,
               "Star Wars: Episode V The Empire Strikes Back": True,
               "Star Wars: Episode VI Return of the Jedi": True
              }
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movies_seen)
star_wars
```

#### Renaming some survey columns

```python
star_wars = star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.": "Seen_1", "Unnamed: 4": "Seen_2", "Unnamed: 5": "Seen_3", "Unnamed: 6": "Seen_4", "Unnamed: 7": "Seen_5", "Unnamed: 8": "Seen_6"})
```

## Cleaning the Ranking Columns

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
                                     "Unnamed: 10": "ranking_2",
                                     "Unnamed: 11": "ranking_3",
                                     "Unnamed: 12": "ranking_4",
                                     "Unnamed: 13": "ranking_5",
                                     "Unnamed: 14": "ranking_6"})
star_wars.head(10)
```

## Finding the Highest-Ranked Movie

```python
rankings = star_wars[star_wars.columns[9:15]].mean()
rankings
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.bar(range(6), rankings)

```

#### The lower the ranking the better (ranged from 1 as most favourite to 6 as least favourite). The top movie is Star Wars: Episode V The Empire Strikes Back. The last 3 movies (Episodes 4-6) are the highest rated, not surpisingly since they are the original movies.

## Finding the Most Viewed Movie

```python
seen = star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6), seen)
```

#### The original movies were seen more often with a large spike after the first original film. The better ranked movies were seen more often.

## Exploring the Data by Binary Segments

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.title("Highest Rated Movies for Males")
```

```python
plt.bar(range(6), females[females.columns[9:15]].mean())
plt.title("Highest Rated Movies for Females")
```
