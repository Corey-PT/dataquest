# Visualizing Earnings Based On College Majors 

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
print (recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
```

### Number of rows in df

```python
raw_data_count = recent_grads.shape
print (raw_data_count[0])
```

### Drop rows with null values

```python
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape
print (cleaned_data_count[0])
```

## Pandas, Scatter Plots

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Median vs Sample Size')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Unemployment Rate vs Sample Size')
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Median vs Full time')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('Unemployment Rate vs Share of Women')
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Median vs Men')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Median vs Women')
```

## All Histograms at once with a for loop

```python
col = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

fig = plt.figure(figsize=(8,16))
for r in range(1,9):
    ax = fig.add_subplot(8,1,r)
    fig.tight_layout()
    ax.set_title(col[r-1])
    ax = recent_grads[col[r-1]].plot(kind='hist')
```
